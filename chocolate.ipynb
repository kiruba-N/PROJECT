# DATA VISUALIZATION

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
chocolates=['Royal','Belgiam','Crunch','Aero','Bliss','Fine','DairyMilk','Dark']
protein=[5,4.5,4.5,4.8,6.9,8.3,7.5,11.6]
```

```python
plt.bar(chocolates,protein,color='red',edgecolor='black')
plt.title('CHOCO NAME')
plt.show()
```

```python
plt.scatter(chocolates,protein,color='black')
plt.title('CHOCO NAME')
plt.show()
```

```python
plt.plot(chocolates,protein,linestyle='dashed')
plt.title('CHOCO NAME')
plt.show()
```

```python
plt.pie(protein,labels=chocolates)
plt.title('CHOCO NAME')
plt.show()
```
